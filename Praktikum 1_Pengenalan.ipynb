```python
import cv2
import matplotlib.pyplot as plt

# Baca Citra
img = cv2.imread("Lenna.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Tampilkan
plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.title("RGB")
plt.subplot(1,2,2); plt.imshow(gray, cmap='gray'); plt.title("Grayscale")
plt.show()

h, w = img.shape[:2]
# Menampilkan tinggi dan lebar gambar
print("Height = {}, Width = {}".format(h,w))

(B, G, R) = img [100, 100]
# Menampilkan jumlah piksel
print("R = {}, G = {}, B = {}".format(R, G, B))

# Histogram RGB & Grayscale
plt.figure(figsize=(10,4))
colors = ('b','r','g')
for i, c in enumerate(colors):
  hist = cv2.calcHist([img], [i], None, [256],[0, 256])
  plt.plot(hist, label=f'Channel {c}')
plt.legend(); plt.title("Histogram RGB"); plt.tight_layout(); plt.show()

plt.figure(); plt.hist(gray.ravel(), bins = 256); plt.title("Histogram Grayscale"); plt.show()
```
